```python
import pandas as pd
import matplotlib.pyplot as plt
log = pd.read_csv('resultsUnet.log', sep=',', engine='python')
log1 = pd.read_csv('resultsResUnetPlus.log', sep=',', engine='python')
log2 = pd.read_csv('resultsMultiResunet.log', sep=',', engine='python')
log3 = pd.read_csv('resultsDeeplabV3plus.log', sep=',', engine='python')
```

```python
ax = plt.gca()
log.plot(kind='line',x='epoch',y='val_loss',ax=ax)
log1.plot(kind='line',x='epoch',y='val_loss', color='red',ax=ax)
log2.plot(kind='line',x='epoch',y='val_loss', color='green', ax=ax)
log3.plot(kind='line',x='epoch',y='val_loss', color='orange', ax=ax)
plt.title('Model Loss Validation')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['U-Net', 'ResUnet++','MultiResUnet','DeepLabV3+'], loc='upper right')
#plt.ylim([0,])
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
from tifffile import imsave
from PIL import Image
import cv2
import datetime, os
```

```python
num_images = 4000
n_classes = 3
print(num_images)
```

```python
image_names = glob.glob("/content/kidStn512_aux/images/*.jpg")
image_names.sort()
image_names_subset = image_names[0:num_images]
images=[]
for image in image_names_subset:
  img = cv2.imread(image, 0)
  redim1 = cv2.resize(img, (256,256))
  images.append(redim1)
image_dataset = np.array(images)
image_dataset = np.expand_dims(image_dataset, axis = 3)
image_dataset = image_dataset /255.
```

```python
mask_names = glob.glob("/content/kidStn512_aux/masks/*.png")
mask_names.sort()
mask_names_subset = mask_names[0:num_images]
masks=[]
for mask in mask_names_subset:
  msk = cv2.imread(mask, 0)
  redim2 = cv2.resize(msk, (256,256))
  masks.append(redim2)
mask_dataset = np.array(masks)
```

```python
mask_dataset_encoded = np.expand_dims(mask_dataset, axis = 3)
print(mask_dataset_encoded.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(image_dataset, mask_dataset_encoded, test_size = 0.2, random_state = 42)
```

```python
import random
test_img_number = random.randint(0, len(X_test)-1)
test_img = X_test[test_img_number]
ground_truth=y_test[test_img_number]
test_img_norm=test_img[:,:,0][:,:,None]
test_img_input=np.expand_dims(test_img_norm, 0)
prediction1 = (model1.predict(test_img_input))
predicted_img1=np.argmax(prediction1, axis=3)[0,:,:]


plt.figure(figsize=(16, 8))
plt.subplot(231)
plt.title('Imagen de Prueba')
plt.imshow(test_img[:,:,0], cmap='gray')
plt.subplot(232)
plt.title('Ground truth')
plt.imshow(ground_truth[:,:,0], cmap='jet')
plt.subplot(233)
plt.title('Prediccion usando U-Net')
plt.imshow(predicted_img1, cmap='jet')
plt.show()
```
